```python
#python program to create a simple calculator

#3 steps to build calculator program
#1.functions for operations
#2.user input
#3.print result

#step 1 create functions
#function to add two numbers
def add(num1,num2):
  return num1+num2

#function to sub two numbers
def sub(num1,num2):
  return num1-num2

#function to multiply two numbers
def multiply(num1,num2):
  return num1*num2

#function to divide two numbers
def divide(num1,num2):
  return num1/num2

#function to avg two numbers
def avg(num1,num2):
  return (num1+num2)/2

#step-2: user input

print("please select a operation:\n" \
      "1.Addition\n"\
      "2.Subtraction\n" \
      "3.multiplication\n" \
      "4.Division\n" \
      "5.Average\n")
select = int(input("select a operation from 1,2,3,4,5:"))

number1=int(input("Enter first number:"))
number2=int(input("Enter second number:"))

#step-3: print result
if select ==1:
  print(number1,"+", number2,"=",\
        add(number1,number2))
elif select ==2:
  print(number1,"-", number2,"=",\
        sub(number1,number2))
elif select==3:
  print(number1,"*",number2,"=",\
        multiply(number1,number2))
elif select==4:
  print(number1,"/",number2,"=",\
        divide(number1,number2))
elif select==5:
  print("(", number1,"+",number2,")","/",2,"=",\
        avg(number1,number2))
else:
  print("invalid operation! please select again!")
```

```python
#4. Arbitrary argument
#Arbitrary positional argument
#Store arguments as TUPLE
def add2number(a,b,):
  return a+b

result=add2number(10,11)
print(result)

def add3number(a,b,c):
  return a+b+c

result1=add3number(10,11,12)
print(result1)

def add_numbers(*args):
  print(type(args))
  return sum(args)

result3=add_numbers(1,2,3,3,5)  #variable no. of arguments
print(result3)

#Arbitrary keyword argument(**Kwargs)
#note: Stores arguments as dictionary

def print_details(**kwargs):
  print(type(kwargs))
  for key, value in kwargs.items():
    print(f"{key}:{value}")

print_details(name="Monika",age=25,city="Mumbai")
```

##String Slicing


```python
my_name="Madhav"
#syntax: string[start:end:step]
print(my_name[0:3])
print(my_name[0:3:1])

print(my_name[0:5:1])
print(my_name[3:5:1])
print(my_name[0:5:2])

print(my_name[0:2])     #first 2 chars
print(my_name[0:3])     #first 3 chars
print(my_name[2:5])     #third to fifth chars
print(my_name[1:4])

print(my_name[-1])      #last char of str
print(my_name[5:])
print(my_name[-2:])
print(my_name[0:5:2])    #every seconf char
print(my_name[:])        #all char
print(my_name[::])       #all char
print(my_name[::-1])     #reverse the string
```
