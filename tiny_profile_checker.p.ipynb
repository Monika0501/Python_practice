```python
Approved={"Excel","Python","Tally","SAP"}

name=input("Name: ").strip()
age=int(input("Age: ").strip())
salary=float(input("Monthly salary: ").strip())
skills=[s.strip().lower() for s in input("skills: ").split(",") if s.strip()]
note=input("Note:").strip()
note=None if note == "" else note

yearly=salary*12
is_adult=age>=18
eligible=is_adult and yearly>300000
has_excel="excel" in skills
has_unknown=any(s not in Approved for s in skills)
note_missing=note is None
days_mod=(age*7) %30


print("\n--- Result ---")
print(f"Hi {name}! Adult? {is_adult}. Yearly ₹{yearly:.2f}. Eligible? {eligible}")
print(f"Skills: {skills} | Has Excel? {has_excel} | Any unknown? {has_unknown}")
print(f"Note missing? {note_missing} | Fun days (age*7 % 30): {days_mod}")
```
