```python
import matplotlib.pyplot as plt
#pyplot is a collection of functions that make matplotlib work as MATLAB
```

##Create first chart

```python
#data
x = [1,2,3,4,5]
y = [10,20,25,35,45]

plt.plot(x,y) #line chart
plt.show() #display chart
```

```python
#customize chart
```

```python
plt.figure(figsize=(4,3))  #figure size
plt.plot(x,y,color='blue',marker='o',linestyle='dashed',linewidth=2,markersize=12)

plt.title("Kuch bhi title hai")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```
